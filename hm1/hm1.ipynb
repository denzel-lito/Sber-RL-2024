```python
import json
from tqdm import tqdm

def solution(mdp, gamma):
    # Заполняем табличку нулями
    q_table = {
        state: {
            action: 0 for action in actions
        } for state, actions in mdp['transition_probs'].items()
    }

    for _ in range(1000):
        for state, actions in mdp['transition_probs'].items():
            for action in actions.keys():
                s = 0
                for next_state, prob in actions[action].items():
                    s+= prob*(mdp['reward_function'][state][action][next_state]+gamma*max(q_table[next_state].values()))
                q_table[state][action] = s

    # selected_actions = {}
    # for state, actions in q_table.items():
    #     selected_actions[state] = max(actions, key=lambda x: q_table[state][x])
    #
    # new_q_table = {}
    # for state, actions in q_table.items():
    return q_table


def main():
    # Список словарей содержит данные о каждом МППР, который определяется
    # состояниями, действиями, вероятностями перехода и вознаграждениями.
    with open('mdps.json', "r") as f:
        mdps = json.load(f)
    # print(mdps[0])

    Qs = []
    for mdp in tqdm(mdps):
        Qs.append(solution(mdp, gamma=0.9))
    with open('submit.json', "w") as f:
        json.dump(Qs, f)
```
